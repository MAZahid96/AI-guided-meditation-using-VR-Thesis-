```python
import pandas as pd
import matplotlib.pyplot as plt

# ---- Load ----
path = r"D:/Git_repository/result folder/survey result/data_project_1069418_2025_12_09.csv"
df = pd.read_csv(path, sep=";", quotechar='"', engine="python")

# ---- Rename participant id + variables ----
df = df.rename(columns={"v_45": "participant_id"})   # your participant id column

rename_map = {
    "v_56": "experience_meditation",
    "v_1":  "condition_meditation_type",   # 1 = conventional, 2 = VR
}
df = df.rename(columns=rename_map)

# ---- Filter valid participants ----
df_valid = df[df["participant_id"].between(100, 143)].copy()

# ---- Keep needed cols ----
data = df_valid[["condition_meditation_type", "experience_meditation"]].copy()

# ---- Clean invalid codes (common SoSci / survey exports) ----
data = data.replace([-77, -99, -66], pd.NA).dropna()

# Ensure numeric
data["experience_meditation"] = pd.to_numeric(data["experience_meditation"], errors="coerce")
data["condition_meditation_type"] = pd.to_numeric(data["condition_meditation_type"], errors="coerce")
data = data.dropna()

# Keep only Likert 1..5
data = data[(data["experience_meditation"] >= 1) & (data["experience_meditation"] <= 5)]
data = data[(data["condition_meditation_type"].isin([1, 2]))]

# ---- Map labels ----
method_map = {1: "Conventional", 2: "VR"}
exp_map = {
    1: "No experience",
    2: "Very little",
    3: "Some",
    4: "Regular",
    5: "Extensive"
}
data["method"] = data["condition_meditation_type"].map(method_map)
data["experience"] = data["experience_meditation"].map(exp_map)

# ---- Count table ----
counts = data.groupby(["method", "experience"]).size().unstack(fill_value=0)

# ---- Plot ----
ax = counts.plot(kind="bar", figsize=(9, 5))
plt.xlabel("Meditation method")
plt.ylabel("Number of participants")
plt.title("Prior meditation experience by meditation method")
plt.legend(title="Prior experience", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Optional: print counts so you can say "16 out of 21" correctly
print(counts)
print("\nTotals per method:\n", data["method"].value_counts())
```

```python
# ----------------------------------------------------
# Helper for Mann–Whitney and nice printing
# ----------------------------------------------------
def mw_test(df, scale, group_var="mode", g1="VR", g2="Conventional", label=""):
    """Run Mann–Whitney U test between two groups and print summary."""
    a = df.loc[df[group_var] == g1, scale].dropna()
    b = df.loc[df[group_var] == g2, scale].dropna()
    
    if len(a) == 0 or len(b) == 0:
        print(f"--- {label} ---")
        print(f"  Not enough valid data for {g1} or {g2} (n={len(a)}, {len(b)}).")
        print()
        return
    
    stat, p = mannwhitneyu(a, b, alternative="two-sided")
    print(f"--- {label} ---")
    print(f"  N_{g1} = {len(a)}, N_{g2} = {len(b)}")
    print(f"  Mann–Whitney U = {stat:.2f}, p = {p:.4f}")
    print(f"  {g1} mean ± SD: {a.mean():.2f} ± {a.std(ddof=1):.2f}")
    print(f"  {g2} mean ± SD: {b.mean():.2f} ± {b.std(ddof=1):.2f}")
    print()

# ----------------------------------------------------
# 1) Baseline perceived stress
# ----------------------------------------------------
mw_test(df, "stress_score",
        group_var="mode",
        g1="VR", g2="Conventional",
        label="Baseline perceived stress (stress_score)")

# ----------------------------------------------------
# 2) Negative affect (pre)
# ----------------------------------------------------
mw_test(df, "affect_pre_neg",
        group_var="mode",
        g1="VR", g2="Conventional",
        label="Negative affect before meditation (affect_pre_neg)")

# ----------------------------------------------------
# 3) State mindfulness during meditation
# ----------------------------------------------------
mw_test(df, "mindfulness_score",
        group_var="mode",
        g1="VR", g2="Conventional",
        label="State mindfulness during meditation (mindfulness_score)")

# ----------------------------------------------------
# 4) Satisfaction with method
# ----------------------------------------------------
mw_test(df, "satisfaction",
        group_var="mode",
        g1="VR", g2="Conventional",
        label="Overall satisfaction with the method (satisfaction)")

# ----------------------------------------------------
# 5) Presence: VR vs Conventional
#     ⚠ For presence, Conventional probably has only missing codes.
#     This test will either fail or show n=0 for Conventional.
#     In the thesis, you will likely report presence only for VR.
# ----------------------------------------------------
mw_test(df, "presence_score",
        group_var="mode",
        g1="VR", g2="Conventional",
        label="Sense of presence in the environment (presence_score)")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# 0) CONFIG
# ----------------------------------------------------------
OUTPUT_DIR = r"D:/Git_repository/result folder/survey result/result plots"
os.makedirs(OUTPUT_DIR, exist_ok=True)

SCALE_LABELS = {
    "stress_score":      "Baseline perceived stress (Likert 1–7)",
    "affect_pre_neg":    "Negative affect, pre-meditation (Likert 1–7)",
    "mindfulness_score": "State mindfulness (Likert 1–7)",
    "presence_score":    "Sense of presence (Likert 1–7)",
    "satisfaction":      "Overall satisfaction (Likert 1–7)",
}

SCALE_RANGES = {
    "stress_score": (1, 7),
    "affect_pre_neg": (1, 7),
    "mindfulness_score": (1, 7),
    "presence_score": (1, 7),
    "satisfaction": (1, 7),
}

MODE_LABELS = {"VR": "VR meditation", "Conventional": "Conventional meditation"}
GROUP_LABELS = {"conventional": "Conventional", "vr_low": "VR (low stress)", "vr_high": "VR (high stress)"}


# ----------------------------------------------------------
# 1) FUNCTIONS
# ----------------------------------------------------------
def boxplot_by_mode(df, scale, filename):
    """VR vs Conventional"""
    vr_vals = df.loc[df["mode"] == "VR", scale].dropna()
    conv_vals = df.loc[df["mode"] == "Conventional", scale].dropna()

    if len(vr_vals) == 0 or len(conv_vals) == 0:
        print(f"[WARN] Not enough data to plot {scale} by mode.")
        return

    fig, ax = plt.subplots(figsize=(4, 5))

    data = [vr_vals, conv_vals]
    labels = [MODE_LABELS["VR"], MODE_LABELS["Conventional"]]

    ax.boxplot(data, tick_labels=labels, showmeans=True, meanline=True)
    ax.set_ylabel(SCALE_LABELS.get(scale, scale))
    if scale in SCALE_RANGES:
        ax.set_ylim(*SCALE_RANGES[scale])

    ax.grid(axis="y", linestyle="--", alpha=0.4)

    out = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out, dpi=300, bbox_inches="tight")
    plt.close()
    print("Saved:", out)


def boxplot_vr_subgroups(df, scale, filename):
    """VR-Low vs VR-High"""
    vr_df = df[df["mode"] == "VR"]

    low_vals = vr_df.loc[vr_df["group"] == "vr_low", scale].dropna()
    high_vals = vr_df.loc[vr_df["group"] == "vr_high", scale].dropna()

    if len(low_vals) == 0 or len(high_vals) == 0:
        print(f"[WARN] Not enough data for VR subgroup plot: {scale}")
        return

    fig, ax = plt.subplots(figsize=(4, 5))
    ax.boxplot([low_vals, high_vals],
               tick_labels=[GROUP_LABELS["vr_low"], GROUP_LABELS["vr_high"]],
               showmeans=True, meanline=True)

    ax.set_ylabel(SCALE_LABELS.get(scale, scale))
    if scale in SCALE_RANGES:
        ax.set_ylim(*SCALE_RANGES[scale])

    ax.grid(axis="y", linestyle="--", alpha=0.4)

    out = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out, dpi=300, bbox_inches="tight")
    plt.close()
    print("Saved:", out)


def single_boxplot_vr(df, scale, filename):
    """Single VR boxplot (Presence)"""
    vals = df.loc[df["mode"] == "VR", scale].dropna()

    if len(vals) == 0:
        print(f"[WARN] No VR data for scale {scale}")
        return

    fig, ax = plt.subplots(figsize=(3.5, 5))
    ax.boxplot([vals], tick_labels=["VR meditation"], showmeans=True, meanline=True)

    ax.set_ylabel(SCALE_LABELS.get(scale, scale))
    if scale in SCALE_RANGES:
        ax.set_ylim(*SCALE_RANGES[scale])

    ax.grid(axis="y", linestyle="--", alpha=0.4)

    out = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out, dpi=300, bbox_inches="tight")
    plt.close()
    print("Saved:", out)


def correlation_heatmap(df, cols, filename):
    """Correlation heatmap for questionnaire scales."""
    data = df[cols].dropna()
    if data.empty:
        print("[WARN] Not enough data for heatmap.")
        return

    corr = data.corr()

    fig, ax = plt.subplots(figsize=(5, 4))
    im = ax.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)

    ax.set_xticks(range(len(cols)))
    ax.set_xticklabels([SCALE_LABELS.get(c, c) for c in cols], rotation=45, ha="right")
    ax.set_yticks(range(len(cols)))
    ax.set_yticklabels([SCALE_LABELS.get(c, c) for c in cols])

    for i in range(len(cols)):
        for j in range(len(cols)):
            ax.text(j, i, f"{corr.iloc[i, j]:.2f}", ha="center", va="center", color="black")

    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    fig.tight_layout()

    out = os.path.join(OUTPUT_DIR, filename)
    fig.savefig(out, dpi=300, bbox_inches="tight")
    plt.close()
    print("Saved:", out)


# ----------------------------------------------------------
# 2) GENERATE ALL FIGURES
# ----------------------------------------------------------

# VR vs Conventional
boxplot_by_mode(df, "stress_score", "fig1_stress_score_VR_vs_Conventional")

boxplot_by_mode(df, "affect_pre_neg",    "fig2_affect_pre_neg_VR_vs_Conventional.png")
boxplot_by_mode(df, "mindfulness_score", "fig3_mindfulness_VR_vs_Conventional.png")
boxplot_by_mode(df, "satisfaction",      "fig4_satisfaction_VR_vs_Conventional.png")

# Presence (VR-only)
single_boxplot_vr(df, "presence_score", "fig5_presence_VR_only.png")

# VR subgroups
boxplot_vr_subgroups(df, "mindfulness_score", "fig6_mindfulness_vr_low_vs_vr_high.png")
boxplot_vr_subgroups(df, "presence_score",    "fig7_presence_vr_low_vs_vr_high.png")
boxplot_vr_subgroups(df, "satisfaction",      "fig8_satisfaction_vr_low_vs_vr_high.png")

# Correlation heatmap
corr_cols = ["stress_score", "affect_pre_neg", "mindfulness_score", "satisfaction"]
correlation_heatmap(df, corr_cols, "fig9_correlation_questionnaire_scales.png")
```
